```python
from sklearn import datasets
import pandas as pd
import numpy as np

diabetes = datasets.load_diabetes(scaled=False)
diabetes_df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)

diabetes_df
```

```python
# Convert sex from 1 2 to Male Female
sexes = []
for i in range(len(diabetes_df)):
    if abs(diabetes_df.iloc[i,1]-1) < 0.1:
        diabetes_df.iloc[i,1] = "Male" 
    else:
        diabetes_df.iloc[i,1] = "Female"
```

```python
# Keep only 4 features
diabetes_4 = diabetes_df.iloc[:, [0,1,2,3]]
diabetes_4
```

```python
# Save to csv
diabetes_4.iloc[:, [0,2,3]] = diabetes_4.iloc[:, [0,2,3]].astype(int)
diabetes_4.to_csv("data/data_ref.csv", index=False)
```

```python
# Add random noise to the data
diabetes_4_noisy = diabetes_4.copy()
diabetes_4_noisy.iloc[:, [0,2,3]] *= np.random.normal(1, 0.02, diabetes_4_noisy.iloc[:, :-1].shape)
diabetes_4_noisy.iloc[:, [0,2,3]] = diabetes_4_noisy.iloc[:, [0,2,3]].astype(int)

# Save to csv
diabetes_4_noisy.to_csv("data/data_noisy.csv", index=False)
```

```python
# Add random noise to the data
diabetes_4_noisy = diabetes_4.copy()
diabetes_4_noisy.iloc[:, [0,2,3]] *= np.random.normal(1, 0.01, diabetes_4_noisy.iloc[:, :-1].shape)
# Convert to int
diabetes_4_noisy.iloc[:, [0,2,3]] = diabetes_4_noisy.iloc[:, [0,2,3]].astype(int)

# Replace some sex with Female
diabetes_4_noisy.iloc[:40, 1] = "Female"
diabetes_4_noisy.to_csv("data/data_female.csv", index=False)
```

```python
# Add random noise to the data
diabetes_4_noisy = diabetes_4.copy()
diabetes_4_noisy.iloc[:, [0,2,3]] *= np.random.normal(1, 0.001, diabetes_4_noisy.iloc[:, :-1].shape)


# Add positive bias to the to columns 2 and 3
diabetes_4_noisy.iloc[:,[2,3]] *= np.random.uniform(1, 1.08, diabetes_4_noisy.iloc[:,[2,3]].shape)
diabetes_4_noisy.iloc[:, [0,2,3]] = diabetes_4_noisy.iloc[:, [0,2,3]].astype(int)

# Save to csv
diabetes_4_noisy.to_csv("data/data_big.csv", index=False)
```
